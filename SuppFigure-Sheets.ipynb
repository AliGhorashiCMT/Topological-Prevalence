```julia
using Crystalline, MPBUtils, JLD2, LinearAlgebra, StaticArrays, SymmetryBases, PyPlot, DelimitedFiles, PrettyTables, Brillouin
using PyCall, Random
np = pyimport("numpy")
include("plotting_utilities.jl")
cm = 1/2.54;
sheets_dir = "./Sheets_of_Lattices/";
```

```julia
sg = 2
id_eps = 1
mode = "te"
output_dir = "./symeigs/output/sg$(sg)/eps$(id_eps)/$(mode)/"
filename = output_dir*"sg$(sg)-epsid$(id_eps)-res64-$(mode)-input.jld2"
loaded_data = load(filename)
flatv = loaded_data["flatv"]
Rsv = loaded_data["Rsv"]; 
isovalv = loaded_data["isovalv"];
```

```julia
idxs = rand(MersenneTwister(1334), 1:10000, 96);
```

```julia
boundariesx = 1
boundariesy = 1
# 8 x 12
fig = figure(figsize=(14*cm, 20*cm))
for (i, idx) in enumerate(idxs)
    subplot(12, 8, i)
    Rs, flat, isoval = Rsv[idx], flatv[idx], isovalv[idx] 
    wigner_cell = wignerseitz(Rs)
    wigner_plot(flat, Rs, wigner_cell; isoval=isoval, fig=fig, ax=gca(), 
        xyz = range(-2, 2, length=500), N=60, in_polygon=true)
    xlim(-boundariesx, boundariesx)
    ylim(-boundariesy, boundariesy)
end
plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0) # In order to get rid of extraneous white space

#savefig(sheets_dir*"sg2-sheets.pdf")
```
